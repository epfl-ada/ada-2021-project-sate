# Share of American speakers  
As one of our research topic is the link between American speakers' characteristics (poltical party, age, genders...) and their opinion about gun owning, it is interesting to compute the share of quotes by Americans in the total number of 'gun quotes'. But most importantly, we create a new column `american` in the dataFrame, to indicates if the speaker is American or not.

```python
df['american'] = (df.nationality.apply(lambda x: (True if 'United States of America' in x else False) if x is not None and len(x) > 0 else False))
df

nb_american = df[(df['american']==True)].shape[0]
print(nb_american)
share_american = nb_american/TOT_LINES_GUNQUOTES * 100
print(share_american)

nb_nationality = df[(df['nationality'].apply(lambda x: (True if x is not None and len(x) > 0 else False)))].shape[0]
share_american_nonone = nb_american/nb_nationality * 100
print(share_american_nonone)

nb_american = df[(df['american']==True)].numOccurrences.sum()
share_american_numOcc = nb_american/TOT_TOT_GUNQUOTES * 100
print(share_american_numOcc)


print('There are ', nb_american, ' quotes said by Americans in the dataFrame.')
print('This represents a share of ', share_american, ' of the dataFrame.')
```
